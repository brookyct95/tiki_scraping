## Introduction
A scraper is a program that visits a Website and reads their page contents to collects certain information.

The main function of our scapper is to collect the product information of all products being sold on [Tiki](https://tiki.vn/). The collected information will be saved as a json file and sent to our Flask app.

The product information includes:
* Product title
* Price
* URL of the product image
* Product Category

For this scraper, we will be using the following libraries:
* **json**: to handle json files
* **requests**: to retrieve the HTML code in our website
* **BeatifulSoup**: to parse the HTML code and get relavant information through HTML tags

```python
# Good luck!
import json
import requests
from bs4 import BeautifulSoup 
```

```python
base_url = "https://tiki.vn/dien-thoai-may-tinh-bang/c1789?src=c.1789.hamburger_menu_fly_out_banner"
response = requests.get(base_url)
print(response)
```

```python
soup=BeautifulSoup(response.text)
print(soup.prettify()[:600])
```

```python
data = soup.find_all('div', class_='product-item')
len(data)
```

```python
print(data[1])
print('--------------------------------------------------------')
print(data[1].text)
```

```python
ids, links,titles, prices, oldprices , images, sales, stars, reviews  =[],[],[],[],[],[],[],[],[]
for i in range(len(data)):
  images.append(data[i].img['src'])
  titles.append(data[i].find(class_="title").text.strip())
  prices.append(data[i].find(class_ = 'final-price').text.strip().split()[0])
  links.append(data[i].find('a')['href'])
  ids.append(data[i].find('a')['data-id'])
  oldprices.append(data[i].find(class_="price-regular").text)
  sales.append(data[i].find(class_="sale-tag sale-tag-square").text)
  stars.append(data[i].find('span', class_ = 'rating-content').span['style'][-3:-1])
  reviews.append(data[i].find(class_='review').text.strip('()'))
```

```python
import json
d = list(zip(ids, links,titles, prices, oldprices , images, sales, stars, reviews))
with open('tikipro.json', 'w') as file:
    json.dump(d, file)
```
